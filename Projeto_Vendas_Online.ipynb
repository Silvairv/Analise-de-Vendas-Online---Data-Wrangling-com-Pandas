```python
import pandas as pd
```

```python
dados = pd.read_json('/content/drive/MyDrive/Arquivos Pandas: transformação e manipulação de dados/dados_vendas_clientes.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[1:])
dados
```

```python
dados.reset_index(drop = True, inplace = True)
dados
```

```python
dados.info()
```

```python
import numpy as np
```

```python
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x : x.replace('R$', '').replace(',', '.').strip())
```

```python
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['Cliente']
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
```

```python
dados['Cliente'] = dados['Cliente'].str.replace('[^a-z ]', '', regex = True)
```

```python
dados['Cliente'] = dados['Cliente'].str.strip()
```

```python
dados.head()
```

```python
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format ='%d/%m/%Y')
```

```python
dados.info()
```

```python
subset = dados.groupby(dados['Cliente'])['Valor da compra'].sum().sort_values(ascending = 0)
subset
```
